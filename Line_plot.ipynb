```python
#LINE PLOT

import matplotlib.pyplot as plt
import numpy as np
 
# define data values
x = np.array([1, 2, 3, 4])  
y = x*2  
# Plot the chart
plt.plot(x, y)  
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")  
plt.title("Basic line chart")
# display
plt.show()  
```

```python
#using pyplot.figure() 
import matplotlib.pyplot as plt
import numpy as np
 
 
x = np.array([1, 2, 3, 4])
y = x*2
 
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("1st graph")
plt.show()  
 
# The figure() function helps in creating new chart.
plt.figure()
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
plt.plot(x1, y1, '-.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2nd graph")
plt.show()
```

```python
#multiple plots

import matplotlib.pyplot as plt
import numpy as np
 
x = np.array([1, 2, 3, 4])
y = x*2
 
# first plot 
plt.plot(x, y)
 
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
 
# second plot 
plt.plot(x1, y1, '-.')
 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('multiple plots')
plt.show()
```

```python
#Filling color between 2 plots
import matplotlib.pyplot as plt
import numpy as np
 
x = np.array([1, 2, 3, 4])
y = x*2
 
plt.plot(x, y)
 
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
 
plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
 
plt.fill_between(x, y, y1, color='pink', alpha=0.5)
plt.show()
```
