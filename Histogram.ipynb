```python
# Basic Histogram
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
X = np.random.randn(1000)
plt.hist(X, bins = 30, color = 'orange',edgecolor = 'black')
plt.xlabel('X-values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
# histogram with density
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

X = np.random.randn(1000)
sns.histplot(X, bins = 30, color = 'green', edgecolor = 'black',kde = True)
plt.xlabel('X-values')
plt.ylabel('Frequency')
plt.title('Histogram with density')
plt.show()
```

```python
#Multiple Histograms with Subplots
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for multiple histograms
data1 = np.random.randn(1000)
data2 = np.random.normal(loc=3, scale=1, size=1000)

# Creating subplots with multiple histograms
fig,axs = plt.subplots(1,2,figsize =(10,10))

# Plot histoframs
axs[0].hist(data1, color = 'yellow',edgecolor = 'black',bins =30)
axs[1].hist(data2, color = 'pink',edgecolor = 'black',bins =30)

# Title and labels
axs[0].set_title('Histogram-1')
axs[1].set_title('Histogram-2')

axs[0].set_xlabel('values')
axs[0].set_ylabel('Frequency')
axs[1].set_xlabel('values')
axs[1].set_ylabel('Frequency')

 
plt.tight_layout()

plt.show()
```

```python
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

np.random.seed(23456789)
n_points = 10000
n_bins = 20

x = np.random.randn(n_points)
y = 0.8 ** x + np.random.randn(10000) + 25 #0.8 raised to x
legend = ['distribution']

fig, ax = plt.subplots(1,1, figsize = (10,7), tight_layout = True)
#if i want to remove spines from histogram
for s in ['top','bottom','left','right']:
    ax.spines[s].set_visible(False)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

ax.grid(color = 'grey', linestyle = '-.', linewidth = 0.5, alpha = 0.6) #alpha is level of transparency -> 0 is transparent
fig.text(0.9,0.15, 'My customized histogram', fontsize = 12, color = 'red', ha='right' , va = 'bottom' , alpha = 0.5 ) #to add text watermark to figure

#Generate histogram
N, bins, patches = ax.hist(x, bins = n_bins)

#for colors -->
fracs = ((N ** (1/5)) / N.max())
norm = colors.Normalize(fracs.min(), fracs.max()) #to generate 
for thisfracs, thispatch in zip(fracs,patches):
    color = plt.cm.viridis(norm(thisfracs)) #viridis -> color map -> visually appealing colors
    thispatch.set_facecolor(color)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(legend)
plt.title('Custom Histogram')
plt.show()
```
